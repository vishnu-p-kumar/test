```python
from transformers import pipeline
qa_pipeline=pipeline("question-answering",model="bert-large-uncased-whole-word-masking-finetuned-squad")
context="""
Apple Inc.is a multinational techology company headquator in Cupertino,California.
It designs,develops,and sells consumer electronics, computer software,ad online services.
Its best-known product inlude the iphone, ipad, and mac computers.
Apple was founded by Steve Jobs, Steve Wozniak, and Wayne in 1976.
"""
print("Ask ad (type 'exit' to stop)")
while True:
    question=input("Enter")
    if question.lower()=="exit":
        print("gf")
        break
    result=qa_pipeline(question=question,context=context)
    print(f"Q:{question}\nA:{result['answer']}\n")
```
