```python
from PIL import Image
img=Image.open("sample.jpg").convert("RGB")
pixels=img.load()
message=input("Enter the hidden text: ")
msg=message+"#"
bits=''.join(format(ord(c),'08b')for c in msg)
max_bits=img.width*img.height
if len(bits)>max_bits:
    raise ValueError("message too long")
i=0
for y in range(img.height):
    for x in range(img.width):
        if i<len(bits):
            r,g,b=pixels[x,y]
            r=(r&~1)|int(bits[i])
            pixels[x,y]=(r,g,b)
            i+=1
        else:
            break
img.save("stego.png")
print("message hidden successfully")
```

```python
from PIL import Image
img=Image.open("stego.png")
pixels=img.load()
bits=""
for y in range(img.height):
    for x in range(img.width):
        bits+=str(pixels[x,y][0]&1)
msg=""
for i in range(0,len(bits),8):
    msg+=chr(int(bits[i:i+8],2))
    if msg.endswith("#"):
        break
print("Hidden message:",msg[:-1])
```
