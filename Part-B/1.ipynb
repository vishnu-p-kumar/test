```python
import numpy as np

def softmax(x):
    e_x = np.exp(x - np.max(x))
    return e_x / e_x.sum()

sequence = np.array([
    [1, 0, 1],
    [0, 2, 0],
    [1, 1, 0]
])

query = np.array([1, 1, 0])

scores = np.dot(sequence, query)

weights = softmax(scores)

context_vector = np.dot(weights, sequence)

print("Input Sequence:\n", sequence)
print("\nQuery:\n", query)
print("\nAttention Scores:\n", scores)
print("\nAttention Weights:\n", weights)
print("\nContext Vector:\n", context_vector)
```
