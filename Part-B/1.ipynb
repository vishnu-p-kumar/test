```python
import numpy as np

def softmax(x):
    e_x = np.exp(x - np.max(x))
    return e_x / e_x.sum()

sequence = np.array([
    [1, 0, 1],
    [0, 2, 0],
    [1, 1, 0]
])

query = np.array([1, 1, 0])

scores = np.dot(sequence, query)

weights = softmax(scores)

context_vector = np.dot(weights, sequence)

print("Input Sequence:\n", sequence)
print("\nQuery:\n", query)
print("\nAttention Scores:\n", scores)
print("\nAttention Weights:\n", weights)
print("\nContext Vector:\n", context_vector)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def softmax(x):
    e_x = np.exp(x - np.max(x))
    return e_x / e_x.sum()

# Words in the sequence
words = ["Apple", "Banana", "Cherry"]

# Vector representation
sequence = np.array([
    [1, 0, 1],
    [0, 2, 0],
    [1, 1, 0]
])

# Query vector
query = np.array([1, 1, 0])

# Attention scores
scores = np.dot(sequence, query)

# Attention weights
weights = softmax(scores)

# Context vector
context_vector = np.dot(weights, sequence)

print("Input Words:", words)
print("\nAttention Scores:", scores)
print("\nAttention Weights:", weights)
print("\nContext Vector:", context_vector)

# ---- Heatmap ----
plt.imshow(weights.reshape(1,-1), cmap='viridis')

plt.xticks(range(len(words)), words)
plt.yticks([])

# Fill values inside graph
for i, value in enumerate(weights):
    plt.text(i, 0, f"{value:.2f}", ha='center', va='center', color='white')

plt.colorbar(label="Attention Weight")
plt.title("Attention Weights Heatmap")

plt.show()
```
