```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

model_name = "google/flan-t5-base"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

question = "If 1 pen costs 10 rupees, what is the cost of 5 pens?"

def generate(prompt):
    inputs = tokenizer(prompt, return_tensors="pt")
    outputs = model.generate(**inputs, max_new_tokens=50)
    return tokenizer.decode(outputs[0], skip_special_tokens=True)

# Zero-shot prompting
zero = generate(f"Answer the question: {question}")

# Few-shot prompting
few = generate("""
Q: If 1 pen costs 10 rupees, what is the cost of 3 pens?
A: 30 rupees

Q: If 1 pencil costs 10 rupees, what is the cost of 4 pencils?
A: 40 rupees

Q: If 1 pen costs 10 rupees, what is the cost of 5 pens?
A:
""")

# Chain-of-thought prompting
cot = generate(f"Solve step by step and give the final answer: {question}")

print("Zero-Shot:", zero)
print("Few-Shot:", few)
print("Chain-of-Thought:", cot)
```

```python
from transformers import pipeline

# Load FLAN-T5 model
generator = pipeline("text2text-generation", model="google/flan-t5-base")

question = "If 1 pen costs 10 rupees, what is the cost of 5 pens?"

def generate(prompt):
    result = generator(prompt, max_new_tokens=50)
    return result[0]['generated_text']

# Zero-shot prompting
zero = generate(f"Answer the question: {question}")

# Few-shot prompting
few = generate("""
Q: If 1 pen costs 10 rupees, what is the cost of 3 pens?
A: 30 rupees

Q: If 1 pencil costs 10 rupees, what is the cost of 4 pencils?
A: 40 rupees

Q: If 1 pen costs 10 rupees, what is the cost of 5 pens?
A:
""")

# Chain-of-thought prompting
cot = generate(f"Solve step by step and give the final answer: {question}")

print("Zero-Shot:", zero)
print("Few-Shot:", few)
print("Chain-of-Thought:", cot)
```
